```python
spot_price = float(gold_price['Close'].iloc[-1])

price_per_gram = spot_price / 31.1035
price_per_baht = spot_price / 15.24

# Calculate for different purities
purities = {
    "GLD9999": 0.9999,  # 24K - 99.99% pure
    "GLD995": 0.9950,   # 99.5% pure
    "GLD965": 0.9650,   # 96.5% pure
    "GLD916": 0.9160,   # 22K - 91.6% pure
    "GLD750": 0.7500,   # 18K - 75% pure
    "GLD585": 0.5850,   # 14K - 58.5% pure
}

print(f"Spot Gold Price: ${spot_price:.2f}/oz")
print(f"Price per gram: ${price_per_gram:.2f}/g")
print(f"Price per baht: ${price_per_baht:.2f}/g")
print("\n" + "="*50)
print("GOLD PRICES BY PURITY (per gram):")
print("="*50)

for code, purity in purities.items():
    purity_price = price_per_gram * purity
    print(f"{code} ({purity*100:.2f}% pure):")
    print(f"  ${purity_price:.2f}/g")
    print(f"  ${purity_price*31.1035:.2f}/oz")
    print(f"  ${purity_price*15.24:.2f}/baht")
    print(f"  Discount from spot: {(1-purity)*100:.1f}%\n")
```

```python
import pandas as pd
from datetime import datetime

# Read all CSV files
files = {
    'gold_price': 'PriceData.csv',
    'volatility': 'Volatility.csv', 
    'tips_yield': 'TipsYield.csv',
    'inflation': 'Inflation.csv',
    'dollar_index': 'DollarIndex.csv'
}

# Initialize with first file
df_combined = None

for name, file in files.items():
    df_temp = pd.read_csv(file)
    
    # Identify the first column (assuming it's the date column)
    date_col = df_temp.columns[0]
    
    # Parse date column, remove timezone info if present
    df_temp['Date'] = pd.to_datetime(df_temp[date_col], errors='coerce')
    
    # Convert all dates to timezone-naive by removing timezone info
    df_temp['Date'] = df_temp['Date'].dt.tz_localize(None)
    
    # Set Date as index and keep only the value column
    # Find value column (the one that's not the original date column)
    value_cols = [col for col in df_temp.columns if col != date_col and col != 'Date']
    
    if len(value_cols) == 0:
        # If only one column exists, use it as value
        value_col = date_col
        df_temp[name] = df_temp[date_col]
    else:
        # Use the first non-date column as value
        value_col = value_cols[0]
        df_temp = df_temp.rename(columns={value_col: name})
    
    # Drop the original date column
    if date_col != 'Date':
        df_temp = df_temp.drop(columns=[date_col])
    
    # Set Date as index
    df_temp = df_temp.set_index('Date')
    
    # Keep only the relevant column
    df_temp = df_temp[[name]]
    
    # Merge with combined dataframe
    if df_combined is None:
        df_combined = df_temp
    else:
        df_combined = df_combined.merge(df_temp, left_index=True, right_index=True, how='outer')

# Reset index if you want Date as a column
df_combined = df_combined.reset_index()

# Sort by date
df_combined = df_combined.sort_values('Date')
df_combined.isnull().sum()
```

```python
# Clean up by first dropping any row with nulled date then interpolate 
# since data is in time-series and fill any data that's still nulled
df_combined = df_combined.dropna(subset=['Date'])

df_interpolated = df_combined.copy()
df_interpolated = df_interpolated.interpolate(method='ffill')
df_interpolated = df_interpolated.set_index('Date')
df_interpolated = df_interpolated.interpolate(method='time')
df_interpolated = df_interpolated.reset_index()

df_filled = df_interpolated.ffill().bfill()
df_filled.isnull().sum()
```

```python
# Save for use (~6k data points with data from 2003~2026)
df_filled.to_csv('CombinedData.csv', index=False)
df_filled.head()
```
